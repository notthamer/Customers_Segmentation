```python
# function to compute hopkins's statistic for the dataframe X
#hopkins used for metric evaluation
from sklearn.neighbors import NearestNeighbors
from random import sample
from numpy.random import uniform
def hopkins(X, ratio=0.05):

    if not isinstance(X, np.ndarray):
      X=X.values  #convert dataframe to a numpy array
    sample_size = int(X.shape[0] * ratio) #0.05 (5%) based on paper by Lawson and Jures

    #a uniform random sample in the original data space
    X_uniform_random_sample = uniform(X.min(axis=0), X.max(axis=0) ,(sample_size , X.shape[1]))

    #a random sample of size sample_size from the original data X
    random_indices=sample(range(0, X.shape[0], 1), sample_size)
    X_sample = X[random_indices]

    #initialise unsupervised learner for implementing neighbor searches
    neigh = NearestNeighbors(n_neighbors=2)
    nbrs=neigh.fit(X)

    #u_distances = nearest neighbour distances from uniform random sample
    u_distances , u_indices = nbrs.kneighbors(X_uniform_random_sample , n_neighbors=2)
    u_distances = u_distances[: , 0] #distance to the first (nearest) neighbour

    #w_distances = nearest neighbour distances from a sample of points from original data X
    w_distances , w_indices = nbrs.kneighbors(X_sample , n_neighbors=2)
    #distance to the second nearest neighbour (as the first neighbour will be the point itself, with distance = 0)
    w_distances = w_distances[: , 1]

    u_sum = np.sum(u_distances)
    w_sum = np.sum(w_distances)

    #compute and return hopkins' statistic
    H = u_sum/ (u_sum + w_sum)
    return H
```

```python
hopkins(df_2, 1)

# With hopkins test, whether the data is uniform or randomly distributed, it returns us a result about it.
# We say that the closer it is to 1, the more prone to data clustering.
```

### inertia

```python
# Initialize an empty list to store the Sum of Squared Distances (SSD) for different numbers of clusters.
ssd = []

# Iterate through a range of cluster numbers from 2 to 10.
for n in range(2, 11):
    # Create a KMeans clustering model with 'n' clusters and set a random seed for reproducibility.
    kmeans = KMeans(n_clusters=n, random_state=42)
    
    # Fit the KMeans model to the standardized data 'X_age_ss'.
    kmeans.fit(X_age_ss)
    
    # Calculate and append the sum of squared distances (inertia) to the 'ssd' list.
    ssd.append(kmeans.inertia_)  # distances from each observation to each cluster centroid

# Create a figure for the plot with a specific size.
plt.figure(figsize=(10, 6))

# Plot the number of clusters on the x-axis and the corresponding SSD on the y-axis.
plt.plot(range(2, 11), ssd, "bo-",  markersize=14.0)

# Set the plot title and labels for the x and y axes.
plt.title('The Elbow Method')
plt.xlabel('No. of Clusters')
plt.ylabel('ssd')

# Display the plot.
plt.show()

# Additional information: You may be interested in reading a new paper about the elbow method.
# The paper can be found at http://arxiv.org/abs/2212.12189
```

### distortion

```python
# Define a range of 'k' values from 2 to 10 for the number of clusters to test.
K = range(2, 11)

# Initialize an empty list to store the distortion values for each 'k'.
distortion = []

# Loop over each 'k' value in the range.
for k in K:
    # Create a KMeans clustering model with 'k' clusters and set a random seed for reproducibility.
    kmeanModel = KMeans(n_clusters=k, random_state=42)
    
    # Fit the KMeans model to the standardized data 'X_age_ss'.
    kmeanModel.fit(X_age_ss)
    
    # Calculate the distances from each observation to each cluster centroid.
    distances = kmeanModel.transform(X_age_ss)
    
    # Get the cluster labels assigned to each data point.
    labels = kmeanModel.labels_
    
    # Initialize an empty list to store cluster-wise distances.
    result = []
    
    # Iterate through each cluster.
    for i in range(k):
        # Extract distances from observations in each cluster to their own centroid.
        cluster_distances = distances[labels == i, i]
        
        # Calculate the mean of squared distances from observations in each cluster to their own centroid and add it to the result list.
        result.append(np.mean(cluster_distances ** 2))
    
    # Calculate the distortion by summing the means of squared distances from all clusters and add it to the distortion list.
    distortion.append(sum(result))

# Create a figure for the plot with a specific size.
plt.figure(figsize=(10, 6))

# Plot the 'k' values on the x-axis and the corresponding distortion values on the y-axis.
plt.plot(K, distortion, "r*--", markersize=14.0)

# Set the x and y axis labels.
plt.xlabel("Different k values")
plt.ylabel("distortion")

# Set the title of the plot.
plt.title("Elbow Method")
```

### silhoutte_score

```python
# Import the necessary function for calculating silhouette scores.
from sklearn.metrics import silhouette_score

# Define a range of 'n_clusters' from 2 to 10 to test different numbers of clusters.
range_n_clusters = range(2, 11)

# Loop over each 'num_clusters' value in the range.
for num_clusters in range_n_clusters:
    # Initialize a KMeans clustering model with 'num_clusters' clusters and set a random seed for reproducibility.
    kmeans = KMeans(n_clusters=num_clusters, random_state=42)
    
    # Fit the KMeans model to the standardized data 'X_age_ss'.
    kmeans.fit(X_age_ss)
    
    # Get cluster labels for each data point.
    cluster_labels = kmeans.labels_
    
    # Calculate the silhouette score, a measure of clustering quality, for the current number of clusters.
    silhouette_avg = silhouette_score(X_age_ss, cluster_labels)
    
    # Print the silhouette score for the current number of clusters.
    print(f"For n_clusters={num_clusters}, the silhouette score is {silhouette_avg}")
```

### Dendrogram

```python
@interact(method=["ward", "complete", "average", "single"])
def dendogramer(method):
    link = linkage(y=X_age_ss, method=method)
    plt.figure(figsize=(20, 10))
    plt.title("Dendogram")
    plt.xlabel("Observations")
    plt.ylabel("Distance")
    dendrogram(link,
               truncate_mode="lastp",
               p=10,
               show_contracted=True,
               leaf_font_size=10)
```
